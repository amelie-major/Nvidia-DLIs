# What is vLLM?

It is an easy-to-deploy Python library for deploying LLMs.

We can:

* Deploy them locally
* Chat to any HuggingFace model
* Even deploy our own fine-tuned models


An **LLM** is a machine learning model that can comprehend and generate human language text, it works by analyzing massive datasets of language. In order to enable the sort of deep learning needed, LLMs are built on neural networks called transformer models. They are able to learn context using a mathematical technique called self-attention to detect subtle ways that elements in a sequence relate to one another. They are only as reliable as the information they ingest but are useful as processing unpredictable queries. 


We can take a single model and serve it with a command, it is powerful and easy to use.
